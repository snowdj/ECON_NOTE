```python
%matplotlib inline 
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random
```

```python
import matplotlib
```

```python
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)
X,Y = np.meshgrid(x, y)
Z = (X**.5 *  Y**.5)

```

```python
fig, ax = plt.subplots()

cnt = ax.contour(Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_wireframe(X, Y, Z, rstride=6, cstride=6, alpha=0.3)

cset = ax.contour(X, Y, Z, zdir='z', offset=0,  cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=0,  cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=2,  cmap=matplotlib.cm.coolwarm)

ax.set_xlim3d(0, 2);
ax.set_ylim3d(0, 2);
ax.set_zlim3d(0, 2);
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1,1,1, projection='3d')

ax.plot_surface(X, Y, Z, rstride=6, cstride=6, alpha=0.1)
cset = ax.contour(X, Y, Z, zdir='z', offset=0,  cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=0,  cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=3,  cmap=matplotlib.cm.coolwarm)

ax.set_xlim3d(0, 2);
ax.set_ylim3d(0, 3);
ax.set_zlim3d(0, 2);
```

```python
fig = plt.figure(figsize=(14,6))

# `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
#ax.view_init(0, 0)


ax.view_init(30, -125)

#ax.view_init(70, 30)


# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

```python
def fun(x, y):
  return x**.5 * y**0.5



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


x = y = np.arange(-0, 6.0, 0.05)

X, Y = np.meshgrid(x, y)

zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])

Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()

```
